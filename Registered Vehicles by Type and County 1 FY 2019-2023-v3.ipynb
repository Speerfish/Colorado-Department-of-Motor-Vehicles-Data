```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the CSV file into a pandas DataFrame
file_name = 'Registered Vehicles by Type and County 1 FY 2019-2023.csv'
df = pd.read_csv(file_name)

# Select the relevant columns for analysis
columns_of_interest = [
    'County', 'Autocycle', 'Passenger', 'Bus', 'Motorcycle', 'Manufactured Home', 
    'Military Historical Vehicle', 'Military Surplus Vehicle', 'Motorhome', 'ATV', 
    'Off Highway Motorcycle', 'Snowmobile', 'Special Mobile Machinery', 
    'Special Truck', 'Special Trailer'
]

# Filter the dataframe to only keep the relevant columns
df_filtered = df[columns_of_interest]

# Display the first few rows of the filtered dataframe to verify the data
df_filtered.head(10)
```

```python
# Set the figure size for better visibility
plt.figure(figsize=(20, 18))

# Loop through each vehicle column and create a bar graph for each one
for column in df_filtered.columns[1:]:  # Skip 'County' for plotting
    plt.figure(figsize=(20, 18))  # Set a figure size for each graph

    # Create a bar plot for the current vehicle type column
    bars = plt.bar(df_filtered['County'], df_filtered[column], color=np.random.rand(len(df_filtered), 3))
    
    # Set title and labels
    plt.title(f'Bar Graph of {column} by County', fontsize=16)
    plt.xlabel('County', fontsize=12)
    plt.ylabel(column, fontsize=12)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=90)
    
    # Add a legend for each county (labelled by County name)
    plt.legend(bars, df_filtered['County'], title='County', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    
    # Adjust layout to accommodate the legend
    plt.tight_layout(rect=[0, 0, 0.85, 1])
    
    # Display the bar plot
    plt.show()
```
